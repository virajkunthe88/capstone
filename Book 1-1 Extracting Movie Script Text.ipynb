```python
import requests
import httplib2
from bs4 import BeautifulSoup, SoupStrainer

#get list of links for all genre pages from the base page
http = httplib2.Http()
status, response = http.request('https://www.actorpoint.com/movie-scripts.html')

genre_links = []

for link in BeautifulSoup(response, 'html.parser',
                             parseOnlyThese=SoupStrainer('a')):
    if link.has_attr('href'):
        if '/movie-scripts/gen-' in link['href'] and link['href'].startswith('https:'):
            genre_links.append(link['href'])

genre_links
```

```python
# function to extract all text given a URL
def get_text(page_url):
    res = requests.get(page_url)
    html_page = res.content
    soup = BeautifulSoup(html_page, 'html.parser')
    text = soup.find_all(text=True)
    script_text = '\n'.join([t for t in text if t.parent.name == 'pre'])
    return script_text
```

```python
# for each genre page, get all the script links and extract the corresponding
# script text and save it to a file with the format 'genre_moviename.txt'

genre_links = [
 'https://www.actorpoint.com/movie-scripts/gen-adventure.html',
 'https://www.actorpoint.com/movie-scripts/gen-animation.html',
 'https://www.actorpoint.com/movie-scripts/gen-comedy.html',
 'https://www.actorpoint.com/movie-scripts/gen-crime.html',
 'https://www.actorpoint.com/movie-scripts/gen-drama.html',
 'https://www.actorpoint.com/movie-scripts/gen-family.html',
 'https://www.actorpoint.com/movie-scripts/gen-fantasy.html',
 'https://www.actorpoint.com/movie-scripts/gen-horror.html',
 'https://www.actorpoint.com/movie-scripts/gen-musical.html',
 'https://www.actorpoint.com/movie-scripts/gen-mystery.html',
 'https://www.actorpoint.com/movie-scripts/gen-romance.html',
 'https://www.actorpoint.com/movie-scripts/gen-science-fiction.html',
 'https://www.actorpoint.com/movie-scripts/gen-thriller.html',
 'https://www.actorpoint.com/movie-scripts/gen-war.html',
 'https://www.actorpoint.com/movie-scripts/gen-western.html']
for link in genre_links:
    genre = link.split('gen-')[1].split(',')[0].split('.')[0]
    http = httplib2.Http()
    status, response = http.request(link)
    for link in BeautifulSoup(response, 'html.parser', parse_only=SoupStrainer('a')):
        if link.has_attr('href'):
            if '/movie-scripts/scripts/' in link['href']:
                script_link = 'https://www.actorpoint.com' + link['href']
                movie_name = script_link.split('/')[-1].split('.')[0]
                text = get_text(script_link)
                filename = 'scripts/' + genre + '_' + movie_name + '.txt'
                print(filename)
                with open(filename, 'w') as f:
                    f.write(text)
```
